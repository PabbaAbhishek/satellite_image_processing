# Visualizing Satellite Imagery with matplotlib

```python
import rasterio
from matplotlib import pyplot as plt
imagery = rasterio.open("4band location.tif")
```

```python
# scaleing values for display purposes
def scale(band):
    return band / 10000.0
blue = scale(imagery.read(1))
green = scale(imagery.read(2))
red = scale(imagery.read(3))
nir = scale(imagery.read(4))
```

```python
display = plt.imshow(blue)
plt.axis('off')
plt.show()
```

```python
display = plt.imshow(green)
display.set_cmap('gist_earth')
plt.axis('off')
plt.show()
```

```python
display = plt.imshow(red)
display.set_cmap('inferno')
plt.axis('off')
plt.show()
```

## Displaying Composite Imagery

```python
import numpy
rgb = numpy.dstack((red, green, blue))
nrg = numpy.dstack((nir,red,green))
```

```python
plt.axis('off')
plt.imshow(rgb)
```

```python
plt.axis('off')
plt.imshow(nrg)
```
